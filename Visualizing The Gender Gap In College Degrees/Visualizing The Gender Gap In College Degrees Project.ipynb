# Visualizing the Gender Gap in College Degrees

In this project, the goal is to visualize the data with the most data/ink ratio and the most meaningful way. 

For the project, I will use the data constituted of degrees and the percent of women bachelors from each degree in the US from 1970 to 2011.

# Introduction

```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Read the dataset
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees
```

# Comparing across STEM Degrees


I will compare men and women distribution in STEM (science, technology, engineering, and mathematics) degrees over time.

```python
# Define two colors for men and women
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

# STEM degrees in the dataset
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

# Plot
fig = plt.figure(figsize=(18, 3))
for sp in range(0,6): # sp stands for subplot
    ax = fig.add_subplot(1,6,sp+1) # add subplot
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False) # to disappear axis lines
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0: # Select first column
        ax.text(2005, 87, 'Men') # add label 'Men' to specified subplot
        ax.text(2002, 8, 'Women') # add label 'Women' to specified subplot
    elif sp == 5: # Select last column
        ax.text(2005, 62, 'Men') 
        ax.text(2001, 35, 'Women')
plt.show()
```

# Comparing across all Degrees

Now, I will compare men and women distribution in all degrees over time.

I will generate a figure having 6x3 shaped plots. In each column of figure, there will be different category of degree groups. These are;

1) STEM
<br>
2) Liberal Arts
<br>
3) Other majors

```python
# Define three colors for men, women and middle line
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (171/255, 171/255, 171/255)

# Create three degree groups in the dataset
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# Plot
fig = plt.figure(figsize=(25, 25))

# First column: STEM Degrees
for sp in range(0,6): # sp stands for subplot
    ax = fig.add_subplot(6,3,3*sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100]) # Only show 0 and 100 tick in y axis
    ax.axhline(50, c=cb_grey, alpha=0.3) # draw horizontal line to indicate 50%
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off') # Turn of tick parameters for all axis and x axis labels
    
    if sp == 0:
        ax.text(2008, 82, 'Women')
        ax.text(2008, 15, 'Men')
    elif sp == 5:
        ax.text(2008, 90, 'Men')
        ax.text(2008, 10, 'Women')
        ax.tick_params(labelbottom='on') # Have x-axis labels in only last subplot

# Second column: Liberal Arts Degrees
for sp in range(0,5): # There are 5 majors in this category not 6
    ax = fig.add_subplot(6,3,3*sp+2) 
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_grey, alpha=0.3)
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    if sp == 0:
        ax.text(2008, 77, 'Women')
        ax.text(2008, 22, 'Men')
    elif sp==4:
        ax.tick_params(labelbottom='on')
        
# Third column: Other Degrees       
for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_grey, alpha=0.3)
    ax.set_title(other_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    if sp == 0: # Select first row
        ax.text(2008, 89, 'Women') 
        ax.text(2008, 7, 'Men')
    elif sp == 5: # Select last row
        ax.text(2008, 62, 'Men')
        ax.text(2008, 36, 'Women')  
        ax.tick_params(labelbottom='on')
        
fig.savefig('gender_degrees.png') # to save the figure
plt.show()
        
```
