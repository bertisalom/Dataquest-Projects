## Pandas, Grouped Bar Plot

 I will compare the number of men with the number of women in each major category using grouped bar plot.

```python
recent_grads['Major_category'].unique()
  
```

```python
# Create dictionary to keep the total number of men and women in each major categories
major_category_men = {}
major_category_women = {}
for major in list(recent_grads['Major_category'].unique()):
    men_count = recent_grads.loc[recent_grads['Major_category'] == major, 'Men'].sum()
    women_count = recent_grads.loc[recent_grads['Major_category'] == major, 'Women'].sum()
    major_category_men[major] = men_count 
    major_category_women[major] = women_count
    
# Create Series from dictionary created above    
men = pd.Series(major_category_men)
women = pd.Series(major_category_women)

# Create DataFrame from Series created above    
maj_cat_men_women = pd.DataFrame(men, columns=['men'])
maj_cat_men_women['women'] = women

# Plot
maj_cat_men_women.plot.bar(figsize=(10,10), title='Number of Men with the Number of Women in Each Major Category')
```

In the majority of major categories, there are more women than men. 

The most crowded major is "Business" with almost equal gender distribution.

The most difference between the number of men and women is in "Education" with women domination. On the other side, "Engineering" is the most men dominated major.
